```python
#use of exceptional handling
#try,except,else,finally

try: #yes or no
   name = input("plz enter your name:" )
   age=int(input("plz enter your age :"))

   print("my name is :",name)
   print("my age is :",age)

except:      #yes error
   print("error is occur")

else:         #no error
   print("congratulation you donot have no error")
   print("no error")

finally:                           #always run
   print("i will be always")

print("i am very important plz execute me !")
```
